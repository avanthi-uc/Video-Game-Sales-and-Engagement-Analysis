```python
cols = ["Times Listed","Number of Reviews", "Plays", "Playing", "Backlogs", "Wishlist"]

for col in cols:
    games[col] = (
        games[col]
        .astype(str)
        .str.lower()
        .apply(lambda x: float(x.replace("k", "")) * 1000 if "k" in x else float(x))
    )
```

```python
games["Team"] = (
    games["Team"]
    .astype(str)
    .str.replace(r"\['", "", regex=True)
    .str.replace(r"'\]", "", regex=True)
    .str.replace(r"\s*,\s*", ",", regex=True)
)
```

```python
games["Reviews"] = (
    games["Reviews"]
    .astype(str)
    .str.replace(r"\['", "", regex=True)
    .str.replace(r"'\]", "", regex=True)
)
```

```python
games["Team"] = games["Team"].str.replace("'", "", regex=False)
games["Reviews"] = games["Reviews"].str.replace("'", "", regex=False)
```

```python
games["Genres"] = (
    games["Genres"]
    .astype(str)
    .str.replace(r"\['", "", regex=True)
    .str.replace(r"'\]", "", regex=True)
    .str.replace("'", "", regex=False)
    .str.replace(r",\s+", ",", regex=True)   # üëà removes space after comma
)
```

```python
games["Reviews"] = (
    games["Reviews"]
    .astype(str)
    .str.replace("[", "", regex=False)
    .str.replace("]", "", regex=False)
    .str.replace("'", "", regex=False)
    .str.replace('"', "", regex=False)
)
```

```python
import pandas as pd
import mysql.connector

# -----------------------
# 1Ô∏è‚É£ Connect to MySQL
# -----------------------
conn = mysql.connector.connect(
    host="localhost",
    port=3306,
    user="root",
    password="root@123"
)

cursor = conn.cursor()

# -----------------------
# 2Ô∏è‚É£ Create Database
# -----------------------
cursor.execute("DROP DATABASE IF EXISTS videogames")
cursor.execute("CREATE DATABASE videogames")
cursor.execute("USE videogames")

# -----------------------
# 3Ô∏è‚É£ Load CSV Files
# -----------------------
sales = pd.read_csv("sales_cleaned.csv")
games = pd.read_csv("games_cleaned.csv")
merged = pd.read_csv("merged.csv")

# -----------------------
# 4Ô∏è‚É£ Function to Auto-Create Table
# -----------------------
def create_table_from_df(df, table_name):
    columns = []
    for col in df.columns:
        columns.append(f"`{col}` TEXT")

    create_table_query = f"""
    CREATE TABLE {table_name} (
        {", ".join(columns)}
    )
    """

    cursor.execute(create_table_query)
    print(f"{table_name} table created.")

# -----------------------
# 5Ô∏è‚É£ Function to Upload Data
# -----------------------
def upload_dataframe(df, table_name):
    cols = ", ".join([f"`{col}`" for col in df.columns])
    placeholders = ", ".join(["%s"] * len(df.columns))
    sql = f"INSERT INTO {table_name} ({cols}) VALUES ({placeholders})"

    data = [tuple(x) for x in df.fillna("").to_numpy()]
    cursor.executemany(sql, data)
    conn.commit()

    print(f"{table_name} uploaded successfully!")

# -----------------------
# 6Ô∏è‚É£ Create Tables
# -----------------------
create_table_from_df(sales, "sales")
create_table_from_df(games, "games")
create_table_from_df(merged, "merged")

# -----------------------
# 7Ô∏è‚É£ Upload Data
# -----------------------
upload_dataframe(sales, "sales")
upload_dataframe(games, "games")
upload_dataframe(merged, "merged")

# -----------------------
# 8Ô∏è‚É£ Close Connection
# -----------------------
cursor.close()
conn.close()

print("ALL DONE ‚úÖ")
```
